```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
cities = ["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"]
bars_in_cities = [8.6, 8.5, 8.3, 7.9, 7.2]
x_axis = np.arange(len(bars_in_cities))
```

```python
# Create a bar chart based upon the above data
plt.bar(x_axis,bars_in_cities,color="r",alpha=0.5, align="center")
#plt.bar(cities,x_axis,color= "b", alpha=0.5, align="center")
```

```python
# Create the ticks for our bar chart's x axis
tick_locations = [values for values in x_axis]
plt.xticks(tick_locations,["New Orleans", "Milwaukee", "Omaha", "Pittsburgh", "Toledo"])
```

```python
# Set the limits of the x axis
plt.xlim(-0.45,len(x_axis)-0.55)
```

```python
# Set the limits of the y axis
plt.ylim(0,max(bars_in_cities)+0.5)
```

```python
# Give the chart a title, x label, and y label
plt.title("Bars in Cities")
plt.xlabel("Cities")
plt.ylabel("NO. of bars")
```

```python
# Save an image of the chart and print it to the screen
plt.savefig("../Images/bar_cities.png")
```
