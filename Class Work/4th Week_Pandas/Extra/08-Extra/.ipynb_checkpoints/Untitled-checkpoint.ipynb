```python
#Read an Excel work sheet from workbook & assign header
df_Students = pd.read_excel(xls,'Student_Mstr',header=None,names=['ID','Name','City','TBD'])

df_Maths = pd.read_excel(xls,'Grades_Math')
df_Eng = pd.read_excel(xls,'Grades_Eng')
print(df_Students)
print(df_Eng.head())
print(df_Maths.head())
```

```python
#del df_Maths['Attempt']
#df_Maths.head()
```

```python
df_Students.drop(['TBD'], axis=1, inplace=True)
df_Students.head(11)
```

```python
df_Students.dropna(how='any').head(15)
```

```python
df_Students.drop([10], axis=0, inplace=True)
```

```python
df_Students.head(15)
```

```python
df_Maths["Subject"]="Maths"
df_Eng["Subject"]="Eng"
print(df_Eng.head())
print(df_Maths.head())
```

```python
df_Maths = df_Maths[["ID","Subject", "Attempt", "MP1", "MP2","MP3", "MP4" ]]
df_Eng = df_Eng[["ID","Subject", "Attempt", "MP1", "MP2","MP3", "MP4" ]]
print(df_Eng.head())
print(df_Maths.head())
```

```python
df_concat = pd.concat([df_Maths,df_Eng],axis=0)
df_concat.head(10)
```

```python
df_concat["Avg_MP_Score"]=df_concat.iloc[:,3:7].mean(axis=1)
df_concat.head()
```

```python
df_concat.groupby(['ID','Subject']).count()
```

```python
new_grp_df = df_concat.groupby(['ID','Subject']).agg({'Attempt':['count'],
                                        'MP1':['max','mean'],'MP2':['min','mean'],
                                        'MP3':['sum','mean'],'MP4':['min','mean'] })
```

```python
new_grp_df.columns = new_grp_df.columns.map('_'.join)
new_grp_df.reset_index()
```

```python
new_grp_df.groupby(['ID']).agg({('Attempt', 'count'):['sum']})
```
